```python
from transformers import pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk
import numpy as np
print("*")
# NLTK'nin stopwords'ları indir
nltk.download('stopwords')
nltk.download('punkt')

# Duygu analizi modelini yükle
sentiment_analysis = pipeline("sentiment-analysis")

# İki metin örneği
text1 = "amerika rusyayı vurdu,hatalıydı"
text2 = "amerika rusyayı silah şatşını engelledi"

# Duygu analizi yap
sentiment1 = sentiment_analysis(text1)[0]['label']
sentiment2 = sentiment_analysis(text2)[0]['label']

# Metinlerin duygularını yazdır
print("Text 1 Sentiment:", sentiment1)
print("Text 2 Sentiment:", sentiment2)

# Konu modelleme için TF-IDF vektörizerini ve LDA modelini hazırla
tfidf_vectorizer = TfidfVectorizer(stop_words=stopwords.words('english'))
lda_model = LatentDirichletAllocation(n_components=5, random_state=42)

# TF-IDF matrisini oluştur
tfidf_matrix = tfidf_vectorizer.fit_transform([text1, text2])

# LDA modelini eğit
lda_model.fit(tfidf_matrix)

# İki metnin LDA temsilini al
lda_representation1 = lda_model.transform(tfidf_vectorizer.transform([text1]))[0]
lda_representation2 = lda_model.transform(tfidf_vectorizer.transform([text2]))[0]

# Cosine Similarity hesapla
cosine_sim_lda = np.dot(lda_representation1, lda_representation2) / (np.linalg.norm(lda_representation1) * np.linalg.norm(lda_representation2))

# Benzerlik oranını yazdır
print("Cosine Similarity (LDA):", cosine_sim_lda)

```
