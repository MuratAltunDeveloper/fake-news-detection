```python
import spacy

# spaCy modelini ykle
nlp = spacy.load("en_core_web_md")

def calculate_semantic_similarity(text1, text2):
    # Metinleri analiz et
    doc1 = nlp(text1)
    doc2 = nlp(text2)

    # Semantik benzerlik hesapla (cosine similarity)
    similarity = doc1.similarity(doc2)
    return similarity

# rnek Kullanm
text1 = "US trade deficit goes up this fiscal"
text2 = "US trade deficit on the rise"

semantic_similarity = calculate_semantic_similarity(text1, text2)

print("Semantic benzerlik:", semantic_similarity)

```
