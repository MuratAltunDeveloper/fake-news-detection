```python
#resmi düzeltme kısmı burada

def improve_image_quality(image_path):
    # Görüntüyü yükle
    img = cv2.imread(image_path)

    # Görüntünün başarıyla yüklendiğini kontrol et
    if img is None:
        print("Görüntü yüklenemedi. Lütfen dosya yolunu kontrol edin.")
        return

    # Görüntüyü gri tonlamaya dönüştür
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Gürültüyü azaltmak için Gaussian bulanıklığı uygula
    blur = cv2.GaussianBlur(gray, (0,0), 3)
    img = cv2.addWeighted(gray, 1.5, blur, -0.5, 0)

    # Görüntüyü yeniden boyutlandır (örneğin, 2 katına çıkar)
    img = cv2.resize(img, None, fx=2, fy=2, interpolation=cv2.INTER_LINEAR)

    # Geliştirilmiş görüntüyü orijinal dosya yoluna kaydet
    cv2.imwrite(image_path, img)

    # Dosya yolunu belirtin
    improve_image_quality('your_image.jpg')
def ocrfonk(dosyayolu):
    # OCR ile metni çıkarma
    image = cv2.imread(dosyayolu)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Gürültüyü azaltmak için Gaussian bulanıklığı
    blur = cv2.GaussianBlur(gray, (3,3), 0)

    # Eşikleme
    _, thresh_img = cv2.threshold(blur, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

    # Gürültüyü kaldırmak için morfolojik operasyonlar
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3,3))
    opening = cv2.morphologyEx(thresh_img, cv2.MORPH_OPEN, kernel, iterations=1)

    # Metni ön plana getirmek için görüntüyü ters çevirme
    invert = 255 - opening

    tessdata_dir_config = r'--tessdata-dir "C:\Program Files\Tesseract-OCR\tessdata"'
    custom_config = r'--oem 3 --psm 6'
    text = pytesseract.image_to_string(invert, lang='eng', config=tessdata_dir_config + custom_config)
    print(f"text: {text}")

def masaustundenAl():
    # Masaüstünden dosya seçme penceresini aç
    file_path = filedialog.askopenfilename(initialdir=r"C:\Users\murat\OneDrive\Masaüstü", title="Masaüstünden Resim Seç", filetypes=(("Image files", "*.jpg; *.jpeg; *.png; *.tif; *.tiff"), ("all files", "*.*")))
    print(f"*********file_path:{file_path}")
    # Seçilen dosyanın yolu eğer boş değilse
    if file_path:
        improve_image_quality(file_path)
        ocrfonk(file_path)

# Ana pencereyi oluştur
root = tk.Tk()
root.title("Fake News Detector")

# Metin girişi (Text Entry) ve Buton ekranına ekleyin
text_entry = tk.Entry(root, width=240)
text_entry.pack(pady=10)

masaustundenAl_button = tk.Button(root, text="Masaüstünden Resim Seç", command=masaustundenAl)
masaustundenAl_button.pack(pady=10)

# Pencere boyutunu belirle
root.geometry("800x600")

# Uygulamayı çalıştır
root.mainloop()
```
