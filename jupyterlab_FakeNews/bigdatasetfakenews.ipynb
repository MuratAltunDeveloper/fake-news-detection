```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.linear_model import SGDClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, classification_report
import pandas as pd
```

```python
#SDG MODEL KULLANILMITIR
# Dosya yollarn belirle
train_file_path =r"C:\Users\murat\OneDrive\Masast\akillilabaratatil\fakenewsdataset2\train.csv"

test_file_path =r"C:\Users\murat\OneDrive\Masast\akillilabaratatil\fakenewsdataset2\test.csv"

# Veri setini ykle
train_data = pd.read_csv(train_file_path)
test_data = pd.read_csv(test_file_path)

# Train ve test veri setlerini ayr
X_train, X_val, y_train, y_val = train_test_split(train_data["text"], train_data["class"], test_size=0.2, random_state=42)

# SGD ile bir metin snflandrma modeli olutur
text_clf = Pipeline([
    ('vect', CountVectorizer()),
    ('tfidf', TfidfTransformer()),
    ('clf', SGDClassifier(loss='hinge', penalty='l2', alpha=1e-3, random_state=42)),
])

# Modeli eit
text_clf.fit(X_train, y_train)

# Modeli deerlendir
y_pred = text_clf.predict(X_val)

# Doruluk skoru ve snflandrma raporu
accuracy = accuracy_score(y_val, y_pred)
report = classification_report(y_val, y_pred)

print(f"Accuracy: {accuracy}")
print("SINIFLANDIRMA N RAPOR:\n", report)
```

```python
# rnek bir metin girii
test_text = "For all their mutual praise, Russian President Vladimir Putin and U.S. President"
#burada metne doru veya yanl demesi datasetteki  verilere bal
# Modeli kullanarak tahmin yap
prediction = text_clf.predict([test_text])

# Tahmin sonucunu yazdr
print("Tahmin Sonucu:", prediction[0])
```
